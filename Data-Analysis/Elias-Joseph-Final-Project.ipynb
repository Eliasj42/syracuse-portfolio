```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from imblearn.over_sampling import SMOTE
import sklearn.metrics
```

```python
X = np.genfromtxt('clean.data.x.csv',delimiter=',')
Y = np.genfromtxt('clean.data.y.csv',delimiter=',')
X = X / X.max(axis=0)
```

```python
sm = SMOTE(random_state = 42)
Xnew, Ynew = sm.fit_resample(X, Y)
Ynew.shape
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(15, activation = 'relu'))
model.add(tf.keras.layers.Dense(250, activation = 'relu'))
model.add(tf.keras.layers.Dense(250, activation = 'relu'))
model.add(tf.keras.layers.Dense(250, activation = 'relu'))
model.add(tf.keras.layers.Dense(2, activation = 'softmax'))

```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(Xnew, to_categorical(Ynew), epochs = 200, batch_size = 10, validation_split = 0.05)
```

```python
sklearn.metrics.precision_score(Y, np.argmax(model.predict(X), axis = 1))
```

```python
sklearn.metrics.recall_score(Y, np.argmax(model.predict(X), axis = 1))
```

```python
sklearn.metrics.accuracy_score(Y, np.argmax(model.predict(X), axis = 1))
```

```python
sklearn.metrics.confusion_matrix(Y, np.argmax(model.predict(X), axis = 1))
```
